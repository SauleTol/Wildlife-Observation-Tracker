```python
import sqlite3
```

```python
conn=sqlite3.connect('wildlife.db')
cursor=conn.cursor()
```

```python
#Created database

cursor.execute("""CREATE TABLE IF NOT EXISTS animals (
                AnimalID INTEGER PRIMARY KEY AUTOINCREMENT,
                CommonName TEXT,
                ScientificName TEXT,
                Habitat TEXT,
                Longevity INTEGER
               )""")

cursor.execute("""CREATE TABLE IF NOT EXISTS sightings (
                SightingID INTEGER PRIMARY KEY AUTOINCREMENT, 
                AnimalID INTEGER,
                LocationID INTEGER, 
                Date DATETYPE,
                Time TIMESTAMP,
                ObserverName TEXT,
                Notes TEXT
               )""")

cursor.execute("""CREATE TABLE IF NOT EXISTS locations (
                LocationID INTEGER PRIMARY KEY AUTOINCREMENT,
                Name TEXT,
                Latitude REAL,
                Longitude REAL,
                Region TEXT
               )""")

conn.commit()
 
```
